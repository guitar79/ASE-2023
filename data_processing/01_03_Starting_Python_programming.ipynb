## 자료형(Data types)



자료형이란 프로그래밍을 할 때 쓰이는 숫자, 문자열 등 자료 형태로 사용하는 모든 것을 뜻한다. 프로그램의 기본이자 핵심 단위가 바로 자료형입니다.

계산 프로그램을 만들려면 어떤 것을 계산할지부터 알아야 하고, 데이터베이스 프로그램을 만들려면 어떤 자료를 저장할지부터 알아야 하는 것처럼 기본 중의 기본이 됩니다. 따라서 자료형을 충분히 이해하지 않고 프로그래밍을 시작하려는 것은 기초 공사가 마무리되지 않은 상태에서 빌딩을 세우는 것과 같습니다.

파이썬 프로그래밍에서는 모든 것이 객체이기 때문에 실제로 데이터 유형은 클래스이고 변수는 이러한 클래스의 인스턴스 (객체) 입니다.

Python에서 사용하는 데이터 타입은 다음과 같은 것들이 있습니니다.

* 없음(None) : NULL 값 개체
* 정수(int)	: 일반 정수 (예 : 32)
* 정수(long)	: 긴 정수. 일반 정수로 쓰기에 너무 긴 정수로서, 접두사 ‘L’과 함께 사용
* 부동소수점수(float)	: 부동소수점수. 10진수 또는 exponent sign을 포함하는 숫자
* 복소수(complex) :복소수. 실수와 허수부의 합으로 된 숫자로 표현
* 부울(Boolean)	: 참(True) 또는 거짓(False) 값 (1과 0의 숫자값으로도 표현)
* 순서(Sequence) : 다음과 같은 유형을 포함: 문자열, 유니코드 문자열, basestring, 리스트, 튜플
* 매핑(Mapping)	: 사전(dictionary) 형을 포함
* 집합(Set)	: 정렬되지 않은 유일한 개체의 모음으로서, 다음과 같은 유형을 포함: set, frozenset
* 파일(File)	: 파일 시스템 개체를 사용할 수 있도록 하는데 사용
* 반복자(Iterator)	: 컨테이너에 대하여 되풀이.

````python
a = 3     # integer(정수)
b = 3.14  # float (실수, real number)
c = 3.14159 # a floating point number
d = 5.1e10 # scientific notation 5.1X10^10
e = 2.5 + 5.3j # a complex number
````

````python
print(a)  # will print 3
print(b)  # will print 3.14
print(c)  
print(d)  
print(e) 
````

````python
hungry = True # boolean literal
need_coffee = False # another boolean literal
````

````python
print(hungry)
print(need_coffee)
````

### 문자열과 메쏘드(methods)

Python에서의 문자열은 일련의 문자로서, 불변(immutable)으로 생성한 후에는 변경할 수 없습니다. 

특정한 문자열의 내용을 조작하는 데 쓰이는 문자열 메소드를 사용할 수 있으며, 이런 메소드들은 문자열을 조작한 사본을 반환해 주고, 원본 문자열은 바뀌지 않게 됩니다. 

메소드(Method)는 클래스, 구조체, 열거형에 포함되어있는 "함수"를 메소드라고 부릅니다. 메소드를 다른말로 "클래스 함수"라고도 합니다. 

Python에 내장된 문자열 메소드 중에서 자주 사용하는 것들은 다음과 같습니다. 

* capitalize() :	문자열의 첫 글자를 대문자로 하는 사본을 반환
* center (width[,fill]) :	지정된 넓이와 선택사항인 채움 문자로 위치가 재조정된 문자열
* endswith(suffix[,start[,end]]) :	문자열이 주어진 형태로 끝나는지의 여부를 반환
* find(sub[,start[,end]]) :	주어진 문자열 조각이 처음 나타나는 곳의 인덱스를 반환
* lower()	: 원본 문자열의 모든 글자를 소문자로 바꾼 사본을 반환
* replace(old,new[,count]) : 원본 문자열에서 old로 주어진 부분을 new로 주어진 부분으로 치환한 사본을 반환
* split([separator[,maxsplit]])	: 문자열 내의 단어 목록을 반환하며 주어진 분리자를 구 분자로 하여 문자열을 왼쪽에서부터 쪼갠다.
* startswith(prefix[,start[,end]])	: 문자열이 주어진 prefix로부터 시작하는지의 여부를 반 환
* upper()	: 문자열의 모든 글자를 대분자로 변환한 문자열 사본을 반환


````python
our_string = 'python is the best language ever'
our_string.capitalize()  # 문자열의 첫 글자를 대문자로
````

하지만 변수의 내용이나 식의 결과를 print 문에 포함하고 싶으면, 다음과 같이 Python의 formatting 옵션을 쓰는 방법이 매우 유용합니다.

서식 옵션은 아래와 같이 방법이 모두 가능하다.

```{admonition} 들여쓰기 내어쓰기
decimalValue, stringValue, floatValue에 들어 있는 값을 출력해 보자.
* (1) print ('String of text goes here', decimalValue, stringValue, floatValue) 
* (2) print ('String of text goes here %d %s %f' % (decimalValue, stringValue, floatValue))
* (3) print ('String of text goes here {0:%d} {1:%s} {2:%f}'.format(decimalValue, stringValue, floatValue)) 
* (4) print (f'String of text goes here {decimalValue} {stringValue} {floatValue}') 
```

* (1)의 방법은 print 문 안에 콤마(,)로 구분하여 변수를 넣어 주는 것입니다.
* (2)의 방법은 텍스트 줄과 변수 목록 사이의 % 기호로 구분하는데, %가 나타내는 것은, 파이썬은 뒤따라오는 변수가 있을 것으로 기대하며, 이러한 변수들의 값은 각자 정해진 위치에서 텍스트의 문자열로 자리잡아야 한다는 것입니다.
* (3)의 방법은 .format( ) 안에 변수를 넣어 줍니다. 
* (4)의 방법은 중괄호({})안체 바로 변수를 넣어 줍니다.

print 문을 사용하여 문자열을 출력할 때에는 많은 선택사항이 있는데, 문자열 서식을 통해 여러 형태로 변환하여 출력할 수 있습니다.

* d	: 부호가 있는 10진수
* i	: 부호가 있는 정수
* o	: 부호 없는 8진수
* u	: 부호 없는 10진수
* x	: 부호 없는 16진수 (소문자)
* X	: 부호없는 16진수 (대문자)
* E	: 부동소수점 지수 형식 (대문자 ‘E’)
* e	: 부동소수점 지수 형식 (소문자 ‘e’)
* f	: 부동소수점 십진수 형식 (소문자)
* F	: 부동소수점 십진수 형식 (‘f’와 동일)
* g	: 지수 < -4 일 경우 부동소수점 지수 형식, 그렇지 않으면 부동소수점수
* G	: 지수 < -4일 경우 부동소수점 지수 형식(대문자), 그렇지 않으면 부동소수점수
* c	: 단일 문자
* r	: 문자열 (임의의 파이썬 개체를 repr()를 사용하여 변환)
* s	: 문자열 (임의의 파이썬 개체를 str()을 사용하여 변환)
* %	: 두번 쓰일 경우, 변환 없이 백분율(%) 문자를 반환


(1)과 (2)는 다음과 같이 적용할 수 있습니다.

````python
x = "Josh"  # 아래와 같이 print문에 문자열을 추가할 수 있다.
y = "Kevin"
print ("My name is", x, ".") # (1)을 사용함.  "," 구분자가 " " 스페이스를 자동 추가함.
print ("My name is %s." % x)  # (2)를 사용함. 
print ("My name is", x, y, ".")  # (1)을 사용함. "," 구분자가 " " 스페이스를 자동 추가함.
print ("My name is %s %s." % (x, y))  # (2)를 사용함. 
````

#### while 문을 강제로 빠져나가

while 문을 강제로 빠져나오고자 할 때는 다음과 같이 조건에 맞는 상황에서 break로 넣어 빠져나올 수 있습니다.




````python
coffee = 10
money = 300
while money:
    print("돈을 받았으니 커피를 줍니다.")
    coffee = coffee -1
    print("남은 커피의 양은 %d개입니다." % coffee)
    if coffee == 0:
        print("커피가 다 떨어졌습니다. 판매를 중지합니다.")
        break
````

#### while문의 맨 처음으로 돌아가기

````python
a = 0
while a < 10:
    a = a + 1
    if a % 2 == 0: continue
    print(a)
````

## Error handling

코드가 오류 조건을 처리 할 수있는 것이 중요하다. 예를 들어, sinc 함수를 구현하고 있을 때 분모에는 0이 들어가서는 안되는데 만약 0을 입력할 경우 다음과 같이 에러가 발생한다.

````python
import math
def sinc(x):
    return math.sin(x) / x
print (sinc(0))
````

이럴때 다음과 같이 에러를 처리할 수 있다. 

````python
def sinc(x):
    try:
        result = math.sin(x) / x
    except ZeroDivisionError:
        result = 1
    return result
print(sinc(0))
print(sinc(0.20))
````
